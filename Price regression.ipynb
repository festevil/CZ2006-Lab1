```python
import pandas as pd
from sklearn import linear_model
```

```python
from statsmodels.formula.api import ols
```

```python
data = pd.read_csv("resale-flat-prices-2017.csv")
df = pd.DataFrame(data)
```

```python
df["remaining_lease"] = df["remaining_lease"].astype(str).str[:2]
df["remaining_lease"] = int(df["remaining_lease"])
df
```

```python
fit = ols('resale_price ~ C(town) + C(flat_type) + C(flat_model) + floor_area_sqm', data=df).fit() 

fit.summary()
```
