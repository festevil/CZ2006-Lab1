### Import the Dataset

The dataset is in CSV format; hence we use the `read_csv` function from Pandas.  
Immediately after importing, take a quick look at the data using the `head` function.

```python
#pre2k=pd.read_csv('approval-date-1990-1999.csv')
post2k=pd.read_csv('approval-date-2000-feb-2012.csv')
post2k.head()

houseTown = pd.DataFrame(post2k['town']) #Town where the house is
houseType = pd.DataFrame(post2k['flat_type']) #Flat type of the house
houseMonth = pd.DataFrame(post2k['month']) #Year in which the resale of the house occurred
housePrice = pd.DataFrame(post2k['resale_price']) #Resale Price of the house in SGD

houseData = pd.DataFrame(post2k['town'], post2k['flat_type'], post2k['month'][0:3], post2k['resale_price'])
```

#### Barchart of House Towns vs Prices

```python
sb.set(rc={'figure.figsize':(11.7,8.27)})
sb.barplot(x='resale_price', y='town', data=post2k)
```

#### Barchart of House Flat Types vs Prices

```python
sb.barplot(x='resale_price', y='flat_type', data=post2k, order=["1 ROOM", "2 ROOM", "3 ROOM", "4 ROOM", "5 ROOM", "EXECUTIVE", "MULTI-GENERATION"])
```
